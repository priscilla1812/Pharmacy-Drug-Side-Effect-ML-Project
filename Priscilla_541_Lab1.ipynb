## Part C: Feature Analysis

1. Identify the most significant features in the dataset.

```python
from sklearn.feature_selection import chi2
from sklearn.preprocessing import LabelEncoder
```

```python
# Encode a few sample features to check relationship with severity target
le = LabelEncoder()
temp_df = pd.DataFrame({
    'drug': le.fit_transform(df['drug_name']),
    'chronic': le.fit_transform(df['chronic_condition']),
    'hospitalized': le.fit_transform(df['hospitalized']),
    'severity': le.fit_transform(df['severity'])
})
chi_scores, p_values = chi2(temp_df[['drug', 'chronic', 'hospitalized']], temp_df['severity'])
print("Chi-Square P-values relative to 'severity' (Lower means more significant):")
for col, p in zip(['drug_name', 'chronic_condition', 'hospitalized'], p_values):
    print(f" - {col}: p = {p:.4f}")
print("\n")
```

```python
# 1. Create a fresh temporary DataFrame for encoding
chi_df = pd.DataFrame()

# 2. Separate and process continuous features by binning them into categories
# (This transforms numerical data into groups so Chi-Square remains statistically valid)
chi_df['age_binned'] = pd.qcut(df['age'], q=4, labels=False, duplicates='drop')
chi_df['dosage_binned'] = pd.qcut(df['dosage_mg'], q=4, labels=False, duplicates='drop')
chi_df['recovery_binned'] = pd.qcut(df['recovery_days'], q=4, labels=False, duplicates='drop')
chi_df['days_to_report_binned'] = pd.qcut(df['days_to_report'], q=4, labels=False, duplicates='drop')

# 3. List of purely categorical/binary features to encode directly
# (Excluding unique IDs like 'patient_id' and raw date columns which aren't features)
categorical_features = [
    'gender', 'country', 'drug_name', 'side_effect', 'outcome',
    'chronic_condition', 'smoker', 'alcohol_use', 'hospitalized', 'age_group'
]

# 4. Apply Label Encoding to categorical features
for col in categorical_features:
    le = LabelEncoder()
    chi_df[col] = le.fit_transform(df[col].astype(str))

# 5. Label Encode the target variable ('severity') separately
le_target = LabelEncoder()
target_encoded = le_target.fit_transform(df['severity'].astype(str))

# 6. Run the Chi-Square test across all processed features
X_chi = chi_df  # All columns are now non-negative integers representing classes
y_chi = target_encoded

chi_scores, p_values = chi2(X_chi, y_chi)

# 7. Compile and display results in a structured DataFrame sorted by significance
chi_results = pd.DataFrame({
    'Feature': X_chi.columns,
    'Chi2 Score': chi_scores,
    'P-Value': p_values
})

# Sort by Chi2 Score descending (Higher Score / Lower P-Value = More Significant)
chi_results = chi_results.sort_values(by='Chi2 Score', ascending=False).reset_index(drop=True)

print("--- Comprehensive Chi-Square Feature Significance Report ---")
print("Interpretation: High Chi2 Score + P-Value < 0.05 implies strong relationship with Severity.\n")
print(chi_results.to_string(formatters={'Chi2 Score': '{:,.2f}'.format, 'P-Value': '{:.4e}'.format}))
```

```python
# Encode 'severity' to numerical for correlation calculation
# The df_temp_corr is not needed if we are going to use df_encoded for correlations
le = LabelEncoder()

# Label Encoding all categorical features (including 'severity')
df_encoded = df.copy()
for col in df_encoded.columns:
    if df_encoded[col].dtype == 'object':
        df_encoded[col] = le.fit_transform(df_encoded[col])

# Calculate correlations on df_encoded which now contains all numerical features
# (original numerical + label-encoded categorical, including 'severity')
correlations = df_encoded.corr(numeric_only=True) # Keep numeric_only=True just in case there are other non-numeric columns we don't want to correlate

# Extract correlations with the 'severity' column (which is now numerical in df_encoded)
# And drop 'severity' itself from the list of correlations
target_corr = correlations['severity'].drop('severity')
target_corr_sorted = target_corr.sort_values(ascending=False)

sns.set(font_scale=0.8)
sns.set_style("white")
sns.set_palette("PuBuGn_d")
sns.heatmap(target_corr_sorted.to_frame(), cmap="YlGnBu", annot=True, fmt='.2f')
plt.title('Correlation with Severity')
plt.show()
```

```python
# heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df_encoded.corr(numeric_only=True), annot=True, cmap='YlGnBu')
plt.title('Correlation Heatmap')
plt.show()
```
